Importing the dependencies

```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# loading the data from the csv file to pandas dataframe
data = pd.read_csv('movies.csv')
```

```python
# printing the first 2 rows of the dataframe
data.head(2)
```

```python
# number of rows and columns in the data frame

data.shape
```

```python
# selecting features for recommendation
features = ['genres','keywords','tagline','cast','director']
```

```python
# replacing the null valuess with null string

for feature in features:
  data[feature] = data[feature].fillna('')
```

```python
# combining all selected features

combined_features = data['genres']+' '+ data['keywords']+' '+ data['tagline']+' '+ data['cast']+' '+ data['director']
```

```python
print(combined_features)
```

```python
# converting the text data to feature vectors

vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

Cosine Similarity

```python
# getting the similarity scores using cosine similarity

similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
# getting movie from user

movie_name = input(' Enter ur favourite movie\'s name : ')
```

```python
# creating a list with all the movie names given in the dataset

list_of_titles = data['title'].tolist()
print(list_of_titles)
```

```python
# finding the close match for the movie name given by the user

find_close_match = difflib.get_close_matches(movie_name, list_of_titles)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
# finding the index of the movie with title

index_of_movie = data[data.title == close_match]['index'].values[0]
print(index_of_movie)
```

```python
# getting a list of similar movies

similarity_score = list(enumerate(similarity[index_of_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
# sorting the movies based on their similarity score

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True) 
print(sorted_similar_movies)
```

```python
# print the name of similar movies based on the index

print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = data[data.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```

Movie Recommendation Sytem

```python
movie_name = input(' Enter ur favourite movie\'s name : ')

list_of_titles = data['title'].tolist()

find_close_match = difflib.get_close_matches(movie_name, list_of_titles)

close_match = find_close_match[0]

index_of_movie = data[data.title == close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_of_movie]))

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True) 

print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = data[data.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```
